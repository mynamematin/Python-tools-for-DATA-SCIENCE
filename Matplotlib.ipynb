# MatPlotLib - Revision

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,8)
y=np.arange(11,19)
```

```python
#plotting using matploylib
##plt scatter
plt.scatter(x,y,c='b')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
```

```python
y=x*x
```

```python
plt.plot(x,y,'r--')
plt.plot(x,y,'b')
```

```python
plt.plot(x,y,'ro--',linestyle='dashed',linewidth=5,markersize=12)
```

```python
##creating subplots
#creating multiple charts in one go
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'ro')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

```python
x=np.arange(0,8*np.pi,0.1)
y=np.sin(x)
plt.title("sine wave form")

#plot the points using matplotlib
plt.plot(x,y)
plt.show()
```

```python
##barplot

x=[2,5,8]
y=[7,9,18]

x2=[4,3,10]
y2=[7,10,13]
plt.bar(x,y,align='center')
plt.bar(x2,y2,color='g')
plt.title('Bar Graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')
```

```python
#box plot
data = [np.random.normal(0,std,100)for std in range(1,4)]
#rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);
```

```python
labels='Python','C++','Ruby','Java'
sizes=[215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0,0,0) #explode 1st slice
#plot
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)

plt.axis('equal')
plt.show()
```
